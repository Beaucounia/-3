```python
import pandas as pd
```

#  Завантажити базу параметрів квітів iris dataset

```python
from sklearn.datasets import load_iris
from sklearn.utils import shuffle
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import r2_score

iris = load_iris()
X = iris.data
y = iris.target
```

# Перемішати записи у завантаженій базі

```python
X, y = shuffle(X, y, random_state=42)
```

# Нормалізація

```python
scaler = StandardScaler()
X_normalized = scaler.fit_transform(X)
```

# Поділ на тестові та начальні вибірки

```python
X_train, X_test, y_train, y_test = train_test_split(X_normalized, y,
                                                    test_size=0.2, random_state=42)
```

# Навчити KNN-регресор з різними значеннями К

```python
result = []
for k in range(2, 20):
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train, y_train)
    y_pred = knn.predict(X_train)
    r2 = r2_score(y_train, y_pred)
    result.append((k, r2))
```

```python
result
```

```python
df = pd.DataFrame(result, columns=['k', 'R^2'])
df
```

# Вибрати величину К для найкращих показників якості регресії

```python
best_k, best_r2 = max(result, key=lambda x: x[1])

print("Найкраще K основано на R2:", best_k)
print("Найкраще R2:", best_r2)
```

# Відтворення на тестовій вибірці

```python
result = []
for k in range(2, 20):
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_test, y_test)
    y_pred = knn.predict(X_test)
    r2 = r2_score(y_test, y_pred)
    result.append((k, r2))
```

```python
df = pd.DataFrame(result, columns=['k', 'R^2'])
df
```

```python
best_k, best_r2 = max(result, key=lambda x: x[1])

print("Найкраще K основано на R2:", best_k)
print("Найкраще R2:", best_r2)
```
